```python
months_list = [m1,m2,m3,m4,m5,m6,m7,m8,m9,m10,m11,m12,m13,m14,m15,m16,m17,m18,m19,m20,m21,m22,m23,m24]  
```

```python
start_id=1
end_id=1
for i in range(len(months_list)):
    if max(months_list[i]['id']) > end_id :
        end_id = max(months_list[i]['id'])
print(start_id," ",end_id)        
```

```python
def combine_all_months():#combine monthwise data into a 3d numpy array [id,month,feature][40000,24,40]
    data = np.zeros([40000,24,40])
    for id_val in np.arange(0,end_id):
        run_ind=0                               
        for month in range(len(months_list)):
            this_month = months_list[month]
            if (id_val+1) in this_month['id'].values :
                data[id_val][run_ind] = this_month[this_month['id']==(id_val+1)].values #append month data to its id,ignore the months which doesnt contain data about this id                
                run_ind += 1
        #data[id_val][run_ind] = np.empty(40).fill(-1)        
        if id_val%4000 == 0:
            print("processed ",id_val)
    return data
            
```

```python
def replace_empty_ref(arr):#replacing the start of invalid data with -1[40](months which dont contain data)
    for i in range(arr.shape[0]):
        refs = list(np.array_equal(x,np.full([40],0)) for x in arr[i])
        if True in (refs):
            ind = refs.index(True)
            arr[i][ind] = np.full([40],-1)
    return arr        
```

```python
merged_data=np.load("../modified_data/merged_months_data.npy")
```

```python
def fit_fun(inp):         #linear regression slope as trend value, ignoring invalid months data 
    if -1 in inp:
        active_months = np.where(inp==-1)[0][0]    #ignore invalid months data
    else:
        active_months = len(inp)
    if active_months==0:
        return 0
        
    x_label = np.arange(1,active_months+1)
    coff=np.polyfit(x_label,inp[0:active_months],1)
    return coff[0]#/((sum(inp))/active_months)
```

```python
trend_data = np.apply_along_axis(fit_fun, 1, merged_data)    #Passes 24 month data for each id and feature and returns trend value. 
```

```python
np.save("../modified_data/trend_data_latest.npy",trend_data)
```
